```python
import PyPDF2

file_1 = "python syallbus.pdf"
file_2 = "maths syallbus.pdf"

def merge_multiple_pages():
    pdf_writer = PyPDF2.PdfWriter()
    
    try:
        # --- PROCESS FILE 1 ---
        reader1 = PyPDF2.PdfReader(file_1)
        pages_to_add_1 = [0, 1] 
        
        for p in pages_to_add_1:
            pdf_writer.add_page(reader1.pages[p])
        
        # --- PROCESS FILE 2 ---
        reader2 = PyPDF2.PdfReader(file_2)
        pages_to_add_2 = [0, 2] 
        
        for p in pages_to_add_2:
            pdf_writer.add_page(reader2.pages[p])

        with open("final_combined_syallbus.pdf", "wb") as f:
            pdf_writer.write(f)
            
        print("Success! Created 'final_combined_syallbus.pdf' with selected pages.")

    except Exception as e:
        print(f"Error: {e}")
merge_multiple_pages()
```
